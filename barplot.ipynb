```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = ('fabric', 'foliage', 'glass', 'leather', 'metal', 'paper', 'plastic', 'stone', 'water', 'wood')
y_pos = np.arange(len(objects)*2,step=2)
fabric_acc = 24 / 25
foliage_acc = 19/19
glass_acc = 24/26
leather_acc = 16/17
metal_acc = 18/20
paper_acc = 23/27
plastic_acc = 28 / 38
stone_acc = 25/26
water_acc = 22/25
wood_acc = 19/25
performance = [fabric_acc,foliage_acc,glass_acc,leather_acc,metal_acc,paper_acc,plastic_acc,stone_acc,water_acc,wood_acc]
barcolors = ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan']
plt.bar(y_pos, performance, align='center', color=barcolors)
plt.xticks(y_pos, objects)
plt.ylabel('Accuracy')
plt.title('Test accuracies by class (Using Xception as base model and FMD test dataset)')

plt.show()
```
